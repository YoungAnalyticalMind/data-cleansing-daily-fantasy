### Split out contest date

```python
contestdate = contests['Contest_Date']
contestdate.head()
```

### Collect all player results in one dataframe

### Get contest standings in one file
split between contest standings and draft rankings

```python
rrcontestdate = contestdate.replace("-", "", regex = True)
rrcontestdateind = rrcontestdate.index.to_series()
# conteststandingslinks = 'drive/data science/data/NBA data/data_warehouse/' + contestdate + '/' + contestdir + '/contest-standings.csv'
conteststandingslinks = '~/google drive/data science/data/NBA data/data_warehouse/' + contestdate + '/' + contestdir + '/contest-standings.csv'
contestdir = contestdir.drop(contestdir.index[[range(290,314)]])
contestdir = contestdir.drop(contestdir.index[[288]])
contestdir = contestdir.reset_index(drop=True)
rrcontestdate = rrcontestdate.drop(rrcontestdate.index[[range(290,314)]])
rrcontestdate = rrcontestdate.drop(rrcontestdate.index[[288]])
rrcontestdate = rrcontestdate.reset_index(drop=True)
conteststandingslinks = conteststandingslinks.drop(conteststandingslinks.index[[range(290,314)]])
conteststandingslinks = conteststandingslinks.drop(conteststandingslinks.index[[288]])
conteststandingslinks = conteststandingslinks.reset_index(drop=True)
```

```python
conteststart = pd.read_csv(conteststandingslinks[0])
conteststart.insert(0,'Contest_Date',rrcontestdate[0])
conteststart.insert(1,'Contest_Name',contestdir[0])
conteststandingsres = conteststart.loc[:,['Contest_Date','Contest_Name','Rank', 'EntryId', 'EntryName', 'Points', 'Lineup']]
contestdraft = conteststart.loc[:,['Contest_Date','Contest_Name','Player','Roster.Position','FPTS','X.Drafted']]
for x in range(len(conteststandingslinks)):
    tempdata = pd.read_csv(conteststandingslinks[x])
    tempdata.insert(0,'Contest_Date',rrcontestdate[x])
    tempdata.insert(1,'Contest_Name',contestdir[x])
    conteststandingsres = conteststandingsres.append(tempdata.loc[:,['Contest_Date','Contest_Name','Rank', 'EntryId', 'EntryName', 'Points', 'Lineup']])
    contestdraft = contestdraft.append(tempdata.loc[:,['Contest_Date','Contest_Name','Player','Roster.Position','FPTS','X.Drafted']])
contestdraft.head()
```

```python
contestdraft = contestdraft.dropna()
```

### Collect payoff structure files 

```python
payoutstruclinks = '~/google drive/data science/data/NBA data/data_warehouse/' + rrcontestdate + '/' + contestdir + '/payout_structure.csv'
payoutvars = 'payout'+ rrcontestdate + rrcontestdateind.astype(str)
payoutvars = payoutvars.drop(payoutvars.index[[range(290,314)]])
payoutvars = payoutvars.drop(payoutvars.index[[288]])
payoutvars = payoutvars.reset_index(drop=True)
payoutstruclinks[1]
```

```python
for x in range(len(payoutvars)):
    try:
        vars()[payoutvars[x]] = pd.read_csv(payoutstruclinks[x])
    except FileNotFoundError:
        vars()[payoutvars[x]] = 'File Does Not Exist'
    if type(vars()[payoutvars[x]]) != str:
        vars()[payoutvars[x]].insert(0,'Contest_Date',rrcontestdate[x])
        vars()[payoutvars[x]].insert(1,'Contest_Name',contestdir[x])
        try:
            payoutres = payoutres.append(vars()[payoutvars[x]])
        except NameError:
            payoutres = vars()[payoutvars[x]]
payoutres.head()
```

### Get projection files in one dataframe 

```python
recontestdate = contestdate.drop_duplicates()
recontestdate = recontestdate.reset_index(drop=True)
rotoprojlinks = '~/google drive/data science/data/NBA data/data_warehouse/projections/rotogrinders/roto_' + recontestdate + '.csv'
recontestdateind = recontestdate.index.to_series()
for x in range(len(rotoprojlinks)):
    try:
        tempdata = pd.read_csv(rotoprojlinks[x])
    except FileNotFoundError:
        tempdata = 'File Not Found'
    if x == 0:
        rotoproj = tempdata
    if type(tempdata) != str:
        tempdata.insert(0,'Contest_Date',recontestdate[x])
        rotoproj = rotoproj.append(tempdata)
rotoproj.head()
```

```python
recontestdate = contestdate.drop_duplicates()
recontestdate = recontestdate.reset_index(drop=True)
dfnprojlinks = '~/google drive/data science/data/NBA data/data_warehouse/projections/dailyfantasynerd/dfn_' + recontestdate + '.csv'
recontestdateind = recontestdate.index.to_series()
tempdata = None
for x in range(len(dfnprojlinks)):
    try:
        tempdata = pd.read_csv(dfnprojlinks[x])
    except FileNotFoundError:
        tempdata = 'File Not Found'
    if x == 0:
        dfnproj = tempdata
    if type(tempdata) != str:
        tempdata.insert(0,'Contest_Date',recontestdate[x])
        dfnproj = dfnproj.append(tempdata)
dfnproj.head()
```
